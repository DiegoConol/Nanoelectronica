### Ejercicio 1

```python
import numpy as np

epsilon = 0.0
t = -1.0

H_pbc = np.array([
    [epsilon, t, 0, 0, 0, t],
    [t, epsilon, t, 0, 0, 0],
    [0, t, epsilon, t, 0, 0],
    [0, 0, t, epsilon, t, 0],
    [0, 0, 0, t, epsilon, t],
    [t, 0, 0, 0, t, epsilon]
], dtype=float)

H_sin_pbc = np.array([
    [epsilon, t, 0, 0, 0, 0],
    [t, epsilon, t, 0, 0, 0],
    [0, t, epsilon, t, 0, 0],
    [0, 0, t, epsilon, t, 0],
    [0, 0, 0, t, epsilon, t],
    [0, 0, 0, 0, t, epsilon]
], dtype=float)

val_pbc, vec_pbc = np.linalg.eigh(H_pbc)
val_sin, vec_sin = np.linalg.eigh(H_sin_pbc)

print("Autovalores con PBCs:")
print(val_pbc)

print("\nAutovectores con PBCs (columnas):")
print(vec_pbc)

print("\nAutovalores sin PBCs:")
print(val_sin)

print("\nAutovectores sin PBCs (columnas):")
print(vec_sin)
```

### Ejercicio 4

```python
import numpy as np
import matplotlib.pyplot as plt

def E_mas(k, eps, t, tp, b=1.0):
    return eps + np.sqrt(t**2 + tp**2 + 2*t*tp*np.cos(k*b))

def E_menos(k, eps, t, tp, b=1.0):
    return eps - np.sqrt(t**2 + tp**2 + 2*t*tp*np.cos(k*b))

# mallado en k
b = 1.0
k = np.linspace(-np.pi/b, np.pi/b, 1000)

# casos del enunciado
casos = [
    (0, 2.0, 2.0),
    (0, 2.0, 1.5),
    (0, 2.0, 1.0),
    (0, 2.0, 0.5),
    (0, 2.0, 0.0),
    (0, 2.0, 2.5),
    (0, 3.0, 2.5),
]

plt.figure(figsize=(12, 10))

for i, (eps, t, tp) in enumerate(casos, start=1):
    plt.subplot(4, 2, i)
    plt.plot(k, E_mas(k, eps, t, tp, b), label=r'$E_+(k)$')
    plt.plot(k, E_menos(k, eps, t, tp, b), label=r'$E_-(k)$')
    plt.axhline(eps, linestyle='--', linewidth=0.8)
    plt.title(rf'$\varepsilon={eps}$ eV, $t={t}$ eV, $t^\prime={tp}$ eV')
    plt.xlabel(r'$k$')
    plt.ylabel('E (eV)')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
